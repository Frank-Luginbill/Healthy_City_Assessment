```python
response=requests.get(url, headers=headers)
data=response.json()
len(data['stops'])
```

```python
bus_stops={
    'distance':[],
    'global_stop_id':[],
    'location_type':[],
    'parent_station_global_stop_id':[],
    'route_type':[],
    'rt_stop_id':[],
    'stop_code':[],
    'stop_lat':[],
    'stop_lon':[],
    'stop_name':[],
    'wheelchair_boarding':[]
}
counter=0
columns=list(bus_stops.keys())
for i in range(len(data['stops'])):
    counter+=1
    for j in range(len(columns)):
        bus_stops[columns[j]].append(data['stops'][i][columns[j]])
```

```python
final_data=pd.DataFrame()
for i in bus_stops.keys():
    final_data[i]=bus_stops[i]
final_data
```
