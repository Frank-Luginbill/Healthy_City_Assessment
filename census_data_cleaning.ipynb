```python
export1=excel.iloc[:,[0,134,138,544]]
export1.columns=['GEO_ID','social security payouts','retirement income','percent below poverty line']
export2=excel2.iloc[:,[0,148]]
export2.columns=['GEO_ID','number insured']
export3=excel3.iloc[:,[0,10,68,76,84]]
export3.columns=['GEO_ID','population of old people','number covered by medicare','number covered by medicaid','number covered by private insurance']
export4=excel4.iloc[:,[0,50,52,54,56,58,60,62,64,66,68]]
export4.columns=['GEO_ID','number disabled','number disabled with insurance','number disabled with with private insurance',
                 'number disabled with medicare','number disabled without insurance',
                'number abled','numbered abled with insurance','number abled with private insurance',
                 'number abled with public insurance','number abled without insurance']
export5=excel5.iloc[:,[0,21,22,23,24,25,26,45,46,47,48,49,50]]
export5.columns=['GEO_ID','number male 65-74','number male 65-74 mental handicap','number male 65-74 no mental handicap',
                'number male 75+','number male 75+ mental handicap','number male 75+ no mental handicap',
                'number female 65-74','number female 65-74 mental handicap','number female 65-74 no mental handicap',
                'number female 75+','number female 75+ mental handicap','number female 75+ no mental handicap']
export7=excel7.iloc[:,[0,46]]
export7.columns=['GEO_ID','number who own households']
export8=excel8.iloc[:,[0,106,108,110,112,114,116,118,120,122,124,126,128,130,132,134,136,138]]
export8.columns=['GEO_ID','number of houses owned','number <10k','10k-15k','15k-20k','20k-25k','25k-30k','30k-35k',
                 '35k-40k','40k-45k','45k-50k','50k-60k','60k-75k','75-100k','100k-125k','125k-150k','150k-200k','200k+']
export9=excel9.iloc[:,[0,4,6,8,10,12]]
export9.columns=['GEO_ID','Number of households that have an old person','number living alone','number living not alone',
                'number living with family','number living NOT with family']

#merging all the things into 1 dataframe
export=pd.merge(export1, export2, on='GEO_ID',how='inner')
export=pd.merge(export, export3, on='GEO_ID',how='inner')
export=pd.merge(export, export4, on='GEO_ID',how='inner')
#export=pd.merge(export, export5, on='GEO_ID',how='inner') 
#export/excel 5 has different GEO_IDs than the other things for some reason so I am excluding it for time's sake
export=pd.merge(export, export7, on='GEO_ID',how='inner')
export=pd.merge(export, export8, on='GEO_ID',how='inner')
export=pd.merge(export, export9, on='GEO_ID',how='inner')
```

```python
def deUSifier(ex):
    holder=ex.find('US')
    if holder >= 0:
        return ex[holder+2:]
    else:
        return ex
```

```python
rename=export['GEO_ID'] #take off everything US and before. AKA only 260650000100, not the 14000000US....
for i in range(len(rename)):
    rename[i]=deUSifier(rename[i])
    
export
```
